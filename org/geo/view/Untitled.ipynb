```python
import matplotlib.pyplot as plt
import numpy as np


from matplotlib.patches import Ellipse, Circle
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig = plt.figure(num='abm', figsize=(7,4))
plt.subplots_adjust(top=1.0, bottom=0, left=0, right=1.0, hspace=0,wspace=0.1)

ax = fig.add_subplot(121)

# ell1 = Ellipse(xy = (0.0, 0.0), width = 4, height = 8, angle = 30.0, facecolor= 'yellow', alpha=0.3)
cir1 = Circle(xy = (0, 1), radius=0.5, alpha=1.0, facecolor='red')
# ax.add_patch(ell1)
# ax.patches(cir1)
ax.add_patch(cir1)

im = ax.imshow(np.random.random(100).reshape((10, 10)))



ax.set_xticks([])
ax.set_yticks([])


ax2 = fig.add_subplot(122)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.spines['left'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['bottom'].set_visible(False)

# Create inset of width 30% and height 40% of the parent axes' bounding box
# at the lower left corner (loc=3)
axins2 = inset_axes(ax2, width="80%", height="30%", loc=4)
axins2.plot(np.random.random(10), np.random.random(10))

axins2 = inset_axes(ax2, width="80%", height="30%", loc=1)
axins2.plot(np.random.random(10), np.random.random(10))

plt.show()
```

```python
def spines_invisible(ax):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
```

```python
import numpy as np
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(7, 7), tight_layout=True)
gs = gridspec.GridSpec(2, 2)

ax_image = fig.add_subplot(gs[:, 0])
ax_image.imshow(np.random.random(100).reshape((10, 10)))
spines_invisible(ax_image)

ax_prob = fig.add_subplot(gs[0, 1])
ax_prob.plot(np.arange(1., 0., -0.1) * 2000., np.arange(1., 0., -0.1))
ax_prob.set_xticks([])
ax_prob.set_ylabel('YLabel1 %d' % 1)

ax_other = fig.add_subplot(gs[1, 1])
ax_other.plot(np.arange(1., 0., -0.1) * 2000., np.arange(1., 0., -0.1))
ax_other.set_xlabel('xLabel1 %d' % 2)
ax_other.set_ylabel('YLabel1 %d' % 2)


fig.align_labels()  # same as fig.align_xlabels(); fig.align_ylabels()

plt.show()
```
