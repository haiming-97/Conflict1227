#  python matplotlib 子图布局

##  1.自定义子图布局

```python
# 四元列表参数：[x,y,width,height]  左下角坐标为(0,0) 右上角坐标为(1,1), 四个参数的值域均为[0,1]
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse, Circle

fig = plt.figure(num='abm', figsize=(8,4))
axes_image = plt.axes([0, 0, 0.5, 0.8])
circle = Circle(xy = (0, 1), radius=0.5, alpha=1.0, facecolor='red')
axes_image.add_patch(circle)
axes_image.imshow(np.random.random(100).reshape((10, 10)))
axes_image.axis('off')

axes_prob1 = plt.axes([0.55, 0.45, 0.3, 0.35])
axes_prob1.set_xticks([])

axes_prob2 = plt.axes([0.55, 0.05, 0.3, 0.35])
axes_prob2.grid(True)
plt.show()
```

## 2. 行列整齐的布局

```python
fig, axes = plt.subplots(2,3,sharex='col',sharey='row')
print(axes)
axes[0,0].text(0.5,0.5,str((0,0)),fontsize=18,ha='center')
plt.show()
```

## 3. 不规则多行多列布局

```python
mean = [0, 0]
cov = [[1, 1], [1, 1]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, wspace=0.5, hspace=0.5)

main_ax = plt.subplot(grid[0:3, 1:4])
plt.plot(x, y, 'ok', markersize=3, alpha=0.2)

y_hist = plt.subplot(grid[0:3, 0], xticklabels=[], sharey=main_ax)
plt.hist(y, 60, orientation='horizontal', color='gray')
y_hist.invert_xaxis() # y轴调换方向

x_hist = plt.subplot(grid[3, 1:4], yticklabels=[], sharex=main_ax)
plt.hist(x, 60, orientation='vertical', color='gray')
x_hist.invert_yaxis()

plt.show()
```
